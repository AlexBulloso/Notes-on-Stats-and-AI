### 1.5 Visualising Relationshisps - 1.6 Saving your plots

```r
ggplot(
    penguins,
    aes(x = species, y = body_mass_g)
) +
    geom_boxplot()
```

```r
ggplot(
    penguins,
    aes(x = body_mass_g, fill = species)
) + 
    geom_histogram(
        mapping = aes(y = after_stat(density)), 
        alpha = 0.2, 
        bins = 30, 
        na.rm = TRUE
        ) +
    geom_density(
        linewidth = 0.75, 
        na.rm = TRUE, 
        mapping = aes(color = species),
        alpha = 0.5
        ) 
```

```r
ggplot(
    penguins,
    aes(x = island, fill = species)
) + 
    geom_bar(position = "fill")
```

```r
ggplot(
    penguins,
    aes(x = flipper_length_mm, y = body_mass_g)
) +
    geom_point(aes(color = species, shape = species)) +
    facet_wrap(~island)
ggsave(filename = "penguin-plot.png", width = 8, height = 3)
```
